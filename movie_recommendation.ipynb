# SISTEMA DE RECOMENDACIÓN DE PELÍCULAS CON ANÁLISIS DE COMPORTAMIENTO

## Introducción
En la era del streaming, plataformas como Netflix, Amazon Prime y Disney+ generan millones de dólares gracias a sus sistemas de recomendación. Estos sistemas no solo mejoran la experiencia del usuario, sino que aumentan el tiempo de visualización y reducen la tasa de abandono.

Este proyecto construye un sistema de recomendación de películas utilizando el dataset MovieLens, aplicando técnicas de filtrado colaborativo y análisis de comportamiento de usuarios. A diferencia de proyectos tradicionales, aquí nos enfocamos en la perspectiva de un Data Analyst: entender por qué los usuarios ven ciertas películas y cómo podemos usar esos patrones para generar recomendaciones valiosas.

## Problemática
Una plataforma de streaming desea implementar un sistema de recomendación que:

1. Ayude a los usuarios a descubrir nuevas películas basadas en sus gustos

2. Aumente el tiempo de visualización en la plataforma

3. Reduzca la "parálisis de elección" (usuarios que pasan 20 minutos eligiendo qué ver)

Actualmente, la plataforma tiene datos históricos de calificaciones pero no sabe cómo aprovecharlos para generar recomendaciones personalizadas.

## Objetivo general
Construir un sistema de recomendación de películas basado en filtrado colaborativo que genere sugerencias personalizadas y permita visualizar el comportamiento de los usuarios a través de un dashboard interactivo.

### Objetivos específicos
1. Analizar el comportamiento de calificación de los usuarios para identificar patrones de consumo.

2. onstruir una matriz usuario-película y calcular la similitud entre películas usando similitud coseno.

3. Desarrollar un sistema de recomendación que sugiera películas similares a partir de un título de entrada.

4. Visualizar los resultados en un dashboard interactivo con Streamlit que permita explorar las recomendaciones.

## Descripción del dataset
Utilizaremos el dataset MovieLens Small (GroupLens Research) , que contiene:

**movies.csv**

`movieId`: Identificador único de la película

`title`: Título de la película

`genres`: Géneros (pipe-separated, ej: "Action|Adventure|Sci-Fi")

**ratings.csv**

`userId`: Identificador único del usuario

`movieId`: Identificador de la película calificada

`rating`: Calificación (0.5 a 5.0, incrementos de 0.5)

`timestamp`: Momento de la calificación

Estadísticas clave:

100,836 calificaciones

9,742 películas

610 usuarios

Escala de calificación: 0.5 - 5.0 estrellas

## Desarrollo


```python
# Descarga del dataset

# Script para descargar MovieLens
import urllib.request
import zipfile
import os

# Crear carpeta data si no existe
os.makedirs('data', exist_ok=True)

# Descargar dataset
url = "http://files.grouplens.org/datasets/movielens/ml-latest-small.zip"
urllib.request.urlretrieve(url, "data/ml-latest-small.zip")

# Extraer archivos
with zipfile.ZipFile("data/ml-latest-small.zip", 'r') as zip_ref:
    zip_ref.extractall("data/")

print("✅ Dataset descargado y extraído")
```
